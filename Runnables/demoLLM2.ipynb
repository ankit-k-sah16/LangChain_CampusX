### Creating a Runnable using the classes we built in the demoLLM file 

```python
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):

    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random
class NakliLLM(Runnable):
    def __init__(self):
        print("LLM Created")

    def invoke(self, prompt):
        response_list=[
            "Tokyo is the capital of Japan.",
            "Samurai are the warrior of Japan.",
            "Fushimi Inari Shrine is a famous shrine in Kyoto, Japan."
            
        ]
        return {'response':random.choice(response_list)}
    
    def predict(self,prompt):
        response_list=[
            "Tokyo is the capital of Japan.",
            "Samurai are the warrior of Japan.",
            "Fushimi Inari Shrine is a famous shrine in Kyoto, Japan."
            
        ]
        return {'response':random.choice(response_list)}
```

```python
class DemoPromptTemplate(Runnable):
    def __init__(self,template,input_variables):
        self.template=template
        self.input_variables=input_variables

    def invoke(self,input_dict):
        return self.template.format(**input_dict)
    
    def format(self,input_dict):
        return self.template.format(**input_dict)
    
```

```python
class demoStrOutputParser(Runnable):
    def __init__(self):
        pass
    
    def invoke(self,input_data):
        return input_data['response']
```

```python
class RunnableConnector(Runnable):

    def __init__(self,runnable_list):
        self.runnable_list=runnable_list

    def invoke(self,input_data):
        for runnable in self.runnable_list:
            input_data=runnable.invoke(input_data)
        return input_data
```

```python
template=DemoPromptTemplate(
    template="Write a {length} poem about {topic}",
    input_variables=['length','topic']
)
```

```python
llm=NakliLLM()
```

```python
parser=demoStrOutputParser()
```

```python
chain=RunnableConnector([template,llm,parser])
chain.invoke({'length':'long','topic':'Japan'})
```

#### Linking a chain to another chain to make a bigger Chain

```python
template1=DemoPromptTemplate(
    template="Write a joke about {topic} ",
    input_variables=['topic']
    )
```

```python
template2=DemoPromptTemplate(
    template="Explain the following joke:{response} ",
    input_variables=['response']
)
```

```python
llm=NakliLLM()
```

```python
parser=demoStrOutputParser()
```

```python

chain1=RunnableConnector([template1,llm])
chain1.invoke({'topic':'Samurai'})
```

```python
chain2=RunnableConnector([template2,llm,parser])
final_chain=RunnableConnector([chain1,chain2])

final_chain.invoke({'topic':'Japan'})
```
