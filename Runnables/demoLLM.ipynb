### Creating a demo LLM with Runnbles-created from scratch

```python
import random
class NakliLLM:
    def __init__(self):
        print("LLM Created")
    
    def predict(self,prompt):
        response_list=[
            "Tokyo is the capital of Japan.",
            "Samurai are the warrior of Japan.",
            "Fushimi Inari Shrine is a famous shrine in Kyoto, Japan."
            
        ]
        return {'response':random.choice(response_list)}
```

```python
llm=NakliLLM()
response=llm.predict("Name a famous shrine in Japan?")
response
```

```python
class DemoPromptTemplate:
    def __init__(self,template,input_variables):
        self.template=template
        self.input_variables=input_variables
    
    def format(self,input_dict):
        return self.template.format(**input_dict)
    
```

```python
template=DemoPromptTemplate(
    template="Write a {length} poem about {topic}",
    input_variables=['length','topic']
)
```

```python
prompt=template.format({"length":'short','topic':'innovation'})
```

```python
llm=NakliLLM()
```

```python
llm.predict(prompt)
```

```python
class DemoLLMChain:
    def __init__(self,llm,prompt):
        self.llm=llm
        self.prompt=prompt
    def run(self,input_dict):
        final_prompt=self.prompt.format(input_dict)
        result=self.llm.predict(final_prompt)

        return result['response']
```

```python
template=DemoPromptTemplate(
    template="Write a {length} poem about {topic}",
    input_variables=['length','topic']
)
```

```python
llm=NakliLLM()
```

```python
chain=DemoLLMChain(llm,template)
```

```python
chain.run({'length':'short','topic':'innovation'})
```
